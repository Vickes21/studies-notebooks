permitido utilizar o ChatGPT!

```python
import sympy as sy
args = [["1001000", "01100111001"],
    ["100", "01111010101"],
    ["011001011", "01111111010"],
    ["1000000", "1101101110010"],
    ["010", "1010011001000"],
    ["100100101", "1101100110100"],
    ["0101111", "1100010001001"],
    ["000", "1100001010110"],
    ["100111101", "0110011100000"],
    ["100110001", "10100001100001"],
    ["000110101", "00010111100001"],
    ["10001010", "10101011111101"],
    ["100000110", "11100000100100"],
    ["001110001", "11101000011111"],
    ["10011011", "10111111000101"],
    ["010101110", "01111001111010"],
    ["101000000", "00010101010110"],
    ["00111101", "10000000010000"]]

results = []
for arg in args:
  page_count = int(arg[0], 2)
  print('Numero da página:', page_count)
  offset = int(arg[1], 2)
  print('Deslocamento:', offset)

  # ACREDITO QUE ESSA FORMULA ESTÁ EQUIVOCADA E DEVERIA SER (pc * ps) + d, ENTRETANTO O PROFESSOR INSTRUI ASSIM:
  # PARA DESCOBRIR O ENDEREÇO FÍSICO É NECESSÁRIO SOMAR O TAMANHO DA PÁGINA AO DESLOCAMENTO E EM SEGUIDA MULTIPLICAR PELO NÚMERO DA PÁGINA
  # SE O NÚMERO DA PAGINA FOR 0, APENAS SOMAR O TAMANHO DA PÁGINA AO DESLOCAMENTO

  page_size = 2**len(arg[1])
  print('Tamanho da página:', page_size)
  physical_address = (page_count * page_size) + offset
  print('Endereço físico:', physical_address)
  print('-----------')

  results.append([page_count, offset, page_size])


# results

```
