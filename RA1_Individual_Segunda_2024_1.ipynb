###**1**
$$\lim_{{x \to 0}} \frac{1 - \cos(x)}{x^2}$$

- Usando a série de Taylor de $cos(x)$ em torno de $x = 0$:

$$\cos(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!} + \cdots$$

- Substituindo isso na expressão original:
$$\lim_{{x \to 0}} \frac{1 - (1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \cdots)}{x^2}$$

- O que simplifica para:
$$\lim_{{x \to 0}} \frac{\frac{x^2}{2!} - \frac{x^4}{4!} + \cdots}{x^2}$$

- Podemos dividir cada termo da série por $x^2$:
$$\lim_{{x \to 0}} \left(\frac{1}{2!} - \frac{x^2}{4!} + \cdots\right)$$

- Agora, à medida que $x \to 0$, todos os termos envolvendo $(x)$ vão para zero, exceto o termo constante $\frac{1}{2!}$.

- Portanto, o limite é:
$$\lim_{{x \to 0}} \frac{1 - \cos(x)}{x^2} = \frac{1}{2!} = \frac{1}{2}$$



```python
import sympy as sy

x, y, h = sy.symbols('x y h')
f=(1-sy.cos(x))/x**2
sy.plot(f,(x,-20,20))
```

- Dentro dos limites analisados, a função é contínua.

###**2**
$$\lim_{{x \to 1}} \frac{\sqrt{x + 8} - 3}{x^2 - 1}$$

- O denominador $x^2 - 1$ é uma diferença de quadrados, que pode ser fatorada como $(x + 1)(x - 1)$.

- O limite quando $x \to 1$ faz o denominador ir a zero, o que sugere uma indeterminação do tipo $\frac{0}{0}$. Para resolver essa indeterminação, podemos multiplicar o numerador e o denominador por seu conjugado:

$$\frac{\sqrt{x + 8} - 3}{x^2 - 1} \cdot \frac{\sqrt{x + 8} + 3}{\sqrt{x + 8} + 3}$$

- Isso nos dá:
$$\frac{x + 8 - 9}{(x + 1)(x - 1)(\sqrt{x + 8} + 3)}$$

- Simplificando, obtemos:
$$\frac{x - 1}{(x + 1)(x - 1)(\sqrt{x + 8} + 3)}$$

- Podemos cancelar $(x - 1)$ do numerador e denominador:
$$\frac{1}{(x + 1)(\sqrt{x + 8} + 3)}$$

- Agora, aplicando o limite $x \to 1$:
$$\lim_{{x \to 1}} \frac{1}{(x + 1)(\sqrt{x + 8} + 3)} = \frac{1}{(1 + 1)(\sqrt{1 + 8} + 3)}$$

- Calculando o valor:
$$\frac{1}{2(3 + 3)} = \frac{1}{12}$$

Portanto, o limite é $\frac{1}{12}$.



Dentro dos limites analisados, a função é contínua e o cálculo do limite confirma a continuidade e o comportamento esperado em $x = 1$.


```python
f = (sy.sqrt(x + 8) - 3) / (x**2 - 1)

sy.plot(f, (x, 0, 2))
```

### **3**

$$\lim_{h \to 0} \frac{(2 + h)^3 - 8}{h}$$



- A expressão $(2 + h)^3 - 8$ pode ser expandida usando a fórmula do binômio de Newton. A expansão de $(2 + h)^3$ é $2^3 + 3 \cdot 2^2h + 3 \cdot 2h^2 + h^3$.

- Substituindo a expansão na expressão original:
$$\lim_{h \to 0} \frac{2^3 + 3 \cdot 2^2h + 3 \cdot 2h^2 + h^3 - 8}{h}$$

- Simplificando a expressão, já que $2^3 - 8 = 0$:
$$\lim_{h \to 0} \frac{12h + 6h^2 + h^3}{h}$$

- Dividindo todos os termos do numerador por $h$:
$$\lim_{h \to 0} (12 + 6h + h^2)$$

- Agora, aplicando o limite quando $h \to 0$:
$$12 + 6 \cdot 0 + 0^2 = 12$$

Portanto, o limite é $12$.


Dentro dos limites analisados, a função é contínua e o cálculo do limite confirma a continuidade e o comportamento esperado em $h = 0$.

```python
f = ((2 + h)**3 - 8) / h
sy.plot(f, (h, -0.5, 0.5))
```

### **4**

$$\lim_{t \to 0} \left( \frac{1}{t\sqrt{t + 1}} - \frac{1}{t} \right)$$

- Encontrar um denominador comum para combinar as frações:
$$\lim_{t \to 0} \left( \frac{1}{t\sqrt{t + 1}} - \frac{1}{t} \right) = \lim_{t \to 0} \left( \frac{\sqrt{t + 1} - 1}{t\sqrt{t + 1}} \right)$$

- Multiplicar o numerador e o denominador pelo conjugado do numerador:
$$\lim_{t \to 0} \left( \frac{\sqrt{t + 1} - 1}{t\sqrt{t + 1}} \cdot \frac{\sqrt{t + 1} + 1}{\sqrt{t + 1} + 1} \right)$$

- Isso nos dá:
$$\lim_{t \to 0} \left( \frac{(t + 1) - 1}{t\sqrt{t + 1}(\sqrt{t + 1} + 1)} \right) = \lim_{t \to 0} \left( \frac{t}{t\sqrt{t + 1}(\sqrt{t + 1} + 1)} \right)$$

- Cancelando $t$ do numerador e do denominador:
$$\lim_{t \to 0} \left( \frac{1}{\sqrt{t + 1}(\sqrt{t + 1} + 1)} \right)$$

- Aplicando o limite quando $t \to 0$:
$$\lim_{t \to 0} \left( \frac{1}{\sqrt{0 + 1}(\sqrt{0 + 1} + 1)} \right) = \frac{1}{1(1 + 1)} = \frac{1}{2}$$

Portanto, o limite é $\frac{1}{2}$.


Dentro dos limites analisados, a função é contínua e o cálculo do limite confirma a continuidade e o comportamento esperado em $t = 0$.

```python
t = sy.symbols('t')
f = (1/(t*sy.sqrt(t + 1))) - (1/t)

sy.plot(f, (t, -10, 10))
```
