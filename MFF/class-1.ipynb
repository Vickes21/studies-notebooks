###Exercício Prático 1
Utilizando o Google Colab, crie um programa em Python que
produza uma imagem (.png) contendo quatro gráficos, dois
de funções contínuas, dois de funções descontínuas. Você
deve escolher as funções e mostrar os gráficos nos intervalos
que mostrem a continuidade, ou a descontinuidade.

```python
import sympy as sy
import numpy as np
x, y, z, i, a = sy.symbols('x y z i a')
expr = x**2
display(expr)
sy.plot(expr)
```

```python
expr = x**2 * sy.sin(x)
display(expr)
sy.plot(expr)
```

```python
expr = sy.Rational(2,3) * sy.sin(x)
display(expr)
sy.plot(expr)
```

```python
expr = 2*x**3
display(expr)
sy.plot(expr)
```

####Funções decontínuas

```python
expr = sy.Heaviside(x)
display(expr)
sy.plot(expr)
expr = sy.Piecewise((1, (0 < x) & (x <= -2)), (x**2, x > 2))
display(expr)
sy.plot(expr)
```
