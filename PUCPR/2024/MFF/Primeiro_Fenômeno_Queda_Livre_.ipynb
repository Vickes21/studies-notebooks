# A cada 5 segundos
Importando os dados encontrados com o dispositivo ajustado para marcar a altura a cada 5 segundos:
Para isso usei a biblioteca: [gspread](https://github.com/burnash/gspread#more-examples)

```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
# Nome da planilha é uma das opções para abrir a planilha
worksheet = gc.open('Dados Projeto 1').sheet1

```

```python
# cria a lista de tempos
tempos = []
# carrega os valores da planilha
for lin in range(3,7):
  tempos.append(worksheet.cell(lin, 1).value)
tempos = [float(x) for x in tempos]
# Exibe a lista de tempos
print(tempos)

# cria a lista de alturas
alturas = []
# carrega os valores da planilha
for lin in range(3,7):
  alturas.append(worksheet.cell(lin, 2).value)
# Exibe a lista de alturas
alturas = [828 - y for y in [float(x) for x in alturas]]
print(alturas) #só para ter certeza
```

## Gráfico das alturas a cada 5 segundos

```python
# criando os arrays
X = np.array(tempos)
Y =  [828-y for y in np.array(alturas)]
plt.grid()
#plt.gca().invert_yaxis()
plt.title("Dados a cada 5 segundos")
plt.xlabel("Tempos")
plt.ylabel("Alturas")
# plotando apenas os pontos
plt.scatter(X,Y)

plt.savefig('acada5segundos.png', dpi=300)
plt.show()
```

Plotando como uma função, a partir destes pontos, teremos:

```python
# plotando em forma de curva
plt.grid()
plt.title("Dados a cada 5 segundos")
plt.xlabel("Tempos")
plt.ylabel("Alturas")
plt.plot(X, Y)
# plotando apenas os pontos
plt.scatter(X,Y)
plt.savefig('acada5segundoslinha.png', dpi=300)
plt.show()
```

# Polinômio de Lagrange
Considere o conjunto de pontos $A=\{(-1,-1),(0,-2),(2,2)\}$ resultante de uma experiência qualquer e encontre uma equação do segundo grau que passe por todos estes pontos.
$$L(x) = \sum_{k=0}^{n-1} \frac{\prod_{j\neq k}(x-x_j )}{\prod_{j\neq k}(x_k-x_j )}$$

```python
x=[-1,0,2]
y=[-1,-2,2]
X = np.arange(-1.5, 2.3, 0.1)
fig = plt.figure(figsize = (10,8))
plt.plot(X, X**2-2, 'b', x, y, 'ro')
plt.title('Polinômio de Lagrange')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

# plotando as coordenadas no gráfico
for i, j in zip(x, y):
   plt.text(i+0.07, j+0.15, '({}, {})'.format(i, j))

plt.savefig('lagrange1.png', dpi=300)
plt.show()
```
