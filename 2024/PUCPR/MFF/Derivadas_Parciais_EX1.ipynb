# Calcule as derivadas parciais listadas as seguir:

Considerando que a derivada parcial de uma função $W$ em relação a $ x $, denotada por $ \frac{\partial Z}{\partial x} $, é calculada tratando $ y $ como uma constante e derivando apenas a parte da função que contém $ x $. A derivada parcial de $W$ em relação a $ y $, denotada por $ \frac{\partial Z}{\partial y} $, é calculada tratando $ x $ como uma constante e derivando apenas a parte da função que contém $ y $.


## Item 1: $$ Z = (6x + 7y)(5x + 3y) $$

### usando expansão de polinômios

Fazendo o produto $$ (6x + 7y)(5x + 3y) $$ teremos:

$$ Z = 30x^2 + 18xy + 35xy + 21y^2 $$
$$ Z = 30x^2 + 53xy + 21y^2 $$

Agora vamos calcular a derivada parcial em relação a $ x $:

$$ \frac{\partial Z}{\partial x} = \frac{\partial}{\partial x} (30x^2 + 53xy + 21y^2) $$
$$ \frac{\partial Z}{\partial x} = 2 \cdot 30x + 53y + 0 $$
$$ \frac{\partial Z}{\partial x} = 60x + 53y $$

E a derivada parcial em relação a $ y $:

$$ \frac{\partial Z}{\partial y} = \frac{\partial}{\partial y} (30x^2 + 53xy + 21y^2) $$
$$ \frac{\partial Z}{\partial y} = 0 + 53x + 2 \cdot 21y $$
$$ \frac{\partial Z}{\partial y} = 53x + 42y $$

Portanto, as derivadas parciais são:

$$ \frac{\partial Z}{\partial x} = 60x + 53y $$
$$ \frac{\partial Z}{\partial y} = 53x + 42y $$

### usando a regra da derivada do produto de funções

$$ u(x,y) = 6x + 7y $$
$$ v(x,y) = 5x + 3y $$

Calculamos as derivadas parciais de $u$ e $v$ primeiro:

$\partial u$
$$ \frac{\partial u}{\partial x} = 6 $$
$$ \frac{\partial u}{\partial y} = 7 $$

$\partial v$
$$ \frac{\partial v}{\partial x} = 5 $$
$$ \frac{\partial v}{\partial y} = 3 $$

A derivada parcial de $W$em relação a $ x $ é:

$$ \frac{\partial Z}{\partial x} = (5x + 3y)(6) + (6x + 7y)(5) $$
$$ \frac{\partial Z}{\partial x} = 30x + 18y + 30x + 35y $$
$$ \frac{\partial Z}{\partial x} = 60x + 53y $$

A derivada parcial de $W$ em relação a $ y $ é:

$$ \frac{\partial Z}{\partial y} = (5x + 3y)(7) + (6x + 7y)(3) $$
$$ \frac{\partial Z}{\partial y} = 35x + 21y + 18x + 21y $$
$$ \frac{\partial Z}{\partial y} = 53x + 42y $$

Assim, as derivadas parciais usando a regra do produto são:

$$ \frac{\partial Z}{\partial x} = 60x + 53y $$
$$ \frac{\partial Z}{\partial y} = 53x + 42y $$


## Item 2 $$ Z = (x^3 + 7y^2)^4 $$

Neste caso, podemos aplicar a regra da cadeia. Definimos $u = x^3 + 7y^2$ e expressamos $Z$ como $ Z = u^4 $. A derivada de $W$ em relação a $ u $ é $ 4u^3 $.

Precisamos agora encontrar o g(x) que será usado para a aplicação da regra da cadeia nas derivadas parcias em relação a $x$ e $y$. Sendo assim já tiramos as derivada parciais de $ u $ em relação a $ x $ e $ y $ são $ u_x = 3x^2 $ e $ u_y = 14y $, respectivamente.

A derivada parcial de $Z$ em relação a $ x $ é dada por:

$$ \frac{\partial Z}{\partial x} = \frac{dZ}{du} \cdot u_x = 4u^3 \cdot 3x^2 = 12x^2(u)^3 = 12x^2(x^3 + 7y^2)^3 $$

E a derivada parcial de $Z$ em relação a $ y $ é dada por:

$$ \frac{\partial Z}{\partial y} = \frac{dZ}{du} \cdot u_y = 4u^3 \cdot 14y = 56y(u)^3 = 56y(x^3 + 7y^2)^3 $$

Portanto, as derivadas parciais de $Z$ são:

$$ \frac{\partial Z}{\partial x} = 12x^2(x^3 + 7y^2)^3 $$
$$ \frac{\partial Z}{\partial y} = 56y(x^3 + 7y^2)^3 $$


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# definindo as variáveis simbólicas e a função
x, y = sp.symbols('x y')
Z = (x**3 + 7*y**2)**4

# calculando as derivadas parciais
dZ_dx_sym = sp.diff(Z, x)
dZ_dy_sym = sp.diff(Z, y)

# para poder usar o numpy e o matplotlib mais fácil
# o numpy é numérico, o sympy simbólico.
dZ_dx = sp.lambdify((x, y), dZ_dx_sym, 'numpy')
dZ_dy = sp.lambdify((x, y), dZ_dy_sym, 'numpy')
Z_num = sp.lambdify((x, y), Z, 'numpy')

x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)

# Calcular os valores de Z e das derivadas parciais na grade
Z_vals = Z_num(X, Y)
dZ_dx_vals = dZ_dx(X, Y)
dZ_dy_vals = dZ_dy(X, Y)

fig, axes = plt.subplots(1, 3, figsize=(18, 5), subplot_kw={'projection': '3d'})

# Gráfico da função Z
axes[0].plot_surface(X, Y, Z_vals, cmap='viridis')
axes[0].set_title('Function $Z = (x^3 + 7y^2)^4$')

# Gráfico da derivada parcial dZ/dx
axes[1].plot_surface(X, Y, dZ_dx_vals, cmap='coolwarm')
axes[1].set_title('Partial Derivative $\partial Z/\partial x$')

# Gráfico da derivada parcial dZ/dy
axes[2].plot_surface(X, Y, dZ_dy_vals, cmap='plasma')
axes[2].set_title('Partial Derivative $\partial Z/\partial y$')


for ax in axes:
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('Value')

plt.tight_layout()
plt.show()
```

## Item 3 $$ Z = x^3 \sin(2y) $$


Para $ \frac{\partial Z}{\partial x} $, temos:

$$
\frac{\partial}{\partial x} (x^3 \sin(2y)) = 3x^2 \sin(2y)
$$

Isso porque a derivada de $ x^3 $ em relação a $ x $ é $ 3x^2 $, e $ \sin(2y) $ é tratado como uma constante durante a diferenciação parcial em relação a $ x $.

Para $ \frac{\partial Z}{\partial y} $, temos:

$$
\frac{\partial}{\partial y} (x^3 \sin(2y)) = x^3 \frac{\partial}{\partial y} (\sin(2y))
$$

Para encontrar a derivada de $ \sin(2y) $ em relação a $ y $, usamos a regra da cadeia:

$$
\frac{\partial}{\partial y} (\sin(2y)) = \cos(2y) \cdot \frac{\partial}{\partial y} (2y) = 2\cos(2y)
$$

Assim, a derivada parcial de $W$ em relação a $ y $ é:

$$
\frac{\partial Z}{\partial y} = x^3 \cdot 2\cos(2y) = 2x^3 \cos(2y)
$$

Portanto, as derivadas parciais de $ Z = x^3 \sin(2y) $ são:

$$
\frac{\partial Z}{\partial x} = 3x^2 \sin(2y)
$$

$$
\frac{\partial Z}{\partial y} = 2x^3 \cos(2y)
$$




```python
#este código foi gerado pelo gpt 4.5
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from mpl_toolkits.mplot3d import Axes3D

# Define as variáveis simbólicas
x, y = sp.symbols('x y')

# Define a função
Z = x**3 * sp.sin(2*y)

# Calcula as derivadas parciais
Zx = sp.diff(Z, x)
Zy = sp.diff(Z, y)

# Converte as expressões sympy para funções numéricas
Z_func = sp.lambdify((x, y), Z, 'numpy')
Zx_func = sp.lambdify((x, y), Zx, 'numpy')
Zy_func = sp.lambdify((x, y), Zy, 'numpy')

# Cria uma grade de valores para x e y
x_vals = np.linspace(-2, 2, 400)
y_vals = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# Calcula os valores de Z, Zx e Zy para cada ponto na grade
Z_vals = Z_func(X, Y)
Zx_vals = Zx_func(X, Y)
Zy_vals = Zy_func(X, Y)

# Plotando a função original e suas derivadas parciais em 3D
fig = plt.figure(figsize=(18, 6))

# Plot da função original
ax = fig.add_subplot(131, projection='3d')
surf = ax.plot_surface(X, Y, Z_vals, cmap='viridis')
fig.colorbar(surf, ax=ax)
ax.set_title('Função Original $Z = x^3 \sin(2y)$')

# Plot da derivada parcial em relação a x
ax = fig.add_subplot(132, projection='3d')
surf = ax.plot_surface(X, Y, Zx_vals, cmap='magma')
fig.colorbar(surf, ax=ax)
ax.set_title('Derivada Parcial em relação a $x$')

# Plot da derivada parcial em relação a y
ax = fig.add_subplot(133, projection='3d')
surf = ax.plot_surface(X, Y, Zy_vals, cmap='coolwarm')
fig.colorbar(surf, ax=ax)
ax.set_title('Derivada Parcial em relação a $y$')

plt.tight_layout()
plt.show()
```

## Item 4 $$ W = \sin(x) \cdot e^y + \cos(z) \cdot x^2 $$

Teremos três derivadas parciais uma em relação a $ x $, outra em relação a $ y $, e mais uma em relação a $ z $.

### Derivada Parcial em relação a $ x $:

Para encontrar a derivada parcial de $W$ em relação a $ x $, tratamos $ y $ e $ z $ como constantes. Usaremos a regra do produto para a primeira parte da função onde $ x $ e $ y $ estão presentes:

$$ \frac{\partial}{\partial x} \left( \sin(x) \cdot e^y \right) = \frac{\partial \sin(x)}{\partial x} \cdot e^y + \sin(x) \cdot \frac{\partial e^y}{\partial x} $$

Como $ e^y $ é constante em relação a $ x $, sua derivada em relação a $ x $ é zero:

$$ \frac{\partial}{\partial x} \left( \sin(x) \cdot e^y \right) = \cos(x) \cdot e^y + \sin(x) \cdot 0 = \cos(x) \cdot e^y $$

Para a segunda parte da função onde $ z $ e $ x $ estão presentes, aplicamos a regra da potência em relação a $ x $:

$$ \frac{\partial}{\partial x} \left( \cos(z) \cdot x^2 \right) = \cos(z) \cdot \frac{\partial x^2}{\partial x} = \cos(z) \cdot 2x $$

Assim, a derivada parcial de $W$ em relação a $ x $ é:

$$ \frac{\partial Z}{\partial x} = \cos(x) \cdot e^y + 2x \cdot \cos(z) $$

### Derivada Parcial em relação a $ y $:

Agora, para derivar em relação a $ y $, consideramos $ x $ e $ z $ como constantes. A função $ e^y $ é a única que contém $ y $:

$$ \frac{\partial}{\partial y} \left( \sin(x) \cdot e^y \right) = \sin(x) \cdot \frac{\partial e^y}{\partial y} $$

A derivada de $ e^y $ em relação a $ y $ é $ e^y $:

$$ \frac{\partial}{\partial y} \left( \sin(x) \cdot e^y \right) = \sin(x) \cdot e^y $$

Como $ \cos(z) \cdot x^2 $ não contém $ y $, sua derivada em relação a $ y $ é zero. Portanto, a derivada parcial de $W$ em relação a $ y $ é:

$$ \frac{\partial Z}{\partial y} = \sin(x) \cdot e^y $$

### Derivada Parcial em relação a $ z $:

Finalmente, para derivar em relação a $ z $, mantemos $ x $ e $ y $ constantes. O termo $ \sin(x) \cdot e^y $ não contém $ z $ e, portanto, sua derivada parcial em relação a $ z $ é zero:

$$ \frac{\partial}{\partial z} \left( \sin(x) \cdot e^y \right) = 0 $$

Apenas o termo $ \cos(z) \cdot x^2 $ contém $ z $:

$$ \frac{\partial}{\partial z} \left( \cos(z) \cdot x^2 \right) = x^2 \cdot \frac{\partial \cos(z)}{\partial z} $$

A derivada de $ \cos(z) $ em relação a $ z $ é $ -\sin(z) $:

$$ \frac{\partial}{\partial z} \left( \cos(z) \cdot x^2 \right) = x^2 \cdot (-\sin(z)) $$

Assim, a derivada parcial de $W$ em relação a $ z $ é:

$$ \frac{\partial Z}{\partial z} = -x^2 \cdot \sin(z) $$

Em resumo, as derivadas parciais de $W$ serão dadas por:

$$ \frac{\partial Z}{\partial x} = \cos(x) \cdot e^y + 2x \cdot \cos(z) $$
$$ \frac{\partial Z}{\partial y} = \sin(x) \cdot e^y $$
$$ \frac{\partial Z}{\partial z} = -x^2 \cdot \sin(z) $$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Definindo as variáveis simbólicas
x, y, z = sp.symbols('x y z')

# aqui temos três variáveis independentes
W = sp.sin(x) * sp.exp(y) + sp.cos(z) * x**2

# as derivadas parciais serão calculadas por
W_x = sp.diff(W, x)
W_y = sp.diff(W, y)
W_z = sp.diff(W, z)

# para compatibilizar com o numpy
W_lamb = sp.lambdify((x, y, z), W, 'numpy')
W_x_lamb = sp.lambdify((x, y, z), W_x, 'numpy')
W_y_lamb = sp.lambdify((x, y, z), W_y, 'numpy')
W_z_lamb = sp.lambdify((x, y, z), W_z, 'numpy')

# não será possível um gráfico em 4D, mas vou usar o mesmo intervalo para todas
# as derivadas parcais
valores = np.linspace(-5, 5, 400)

# Para simplificar, consideramos y = 1 e z = 1 e variamos apenas o x
# então veremos sempre um curva mostrando a variação de um dos componentes
W_valores = W_lamb(valores, 1, 1)
W_x_valores = W_x_lamb(valores, 1, 1)
W_y_valores = W_y_lamb(valores, 1, 1)
W_z_valores = W_z_lamb(valores, 1, 1)

# Plotando a função e suas derivadas parciais
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(valores, W_valores, label='W(x, 1, 1)')
plt.title('Função W')
plt.xlabel('x')
plt.ylabel('W')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(valores, W_x_valores, label='∂W/∂x', color='orange')
plt.title('Derivada parcial de W em relação a x')
plt.xlabel('x')
plt.ylabel('∂W/∂x')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(valores, W_y_valores, label='∂W/∂y', color='green')
plt.title('Derivada parcial de W em relação a y')
plt.xlabel('x')  # y é constante
plt.ylabel('∂W/∂y')
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(valores, W_z_valores, label='∂W/∂z', color='red')
plt.title('Derivada parcial de W em relação a z')
plt.xlabel('x')  # z é constante
plt.ylabel('∂W/∂z')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## Item 5 $$ W = e^{-x} \cdot \tan(y) - y \cdot \ln(z) $$


A derivada parcial de $ W $ em relação a $ x $ é:

$$
\frac{\partial W}{\partial x} = \frac{\partial}{\partial x} \left( e^{-x} \cdot \tan(y) \right) - \frac{\partial}{\partial x} \left( y \cdot \ln(z) \right)
$$

A derivada de $ e^{-x} $ em relação a $ x $ é $ -e^{-x} $, enquanto a derivada de $ y \cdot \ln(z) $ em relação a $ x $ é zero, pois não há $ x $ em $ y \cdot \ln(z) $. Assim, a derivada parcial em relação a $ x $ é:

$$
\frac{\partial W}{\partial x} = -e^{-x} \cdot \tan(y)
$$

Agora, para a derivada parcial de $ W $ em relação a $ y $:

$$
\frac{\partial W}{\partial y} = \frac{\partial}{\partial y} \left( e^{-x} \cdot \tan(y) \right) - \frac{\partial}{\partial y} \left( y \cdot \ln(z) \right)
$$

A derivada de $ \tan(y) $ em relação a $ y $ é $ \sec^2(y) $, e a derivada de $ y $ em relação a $ y $ é $ 1 $, então a derivada parcial em relação a $ y $ é:

$$
\frac{\partial W}{\partial y} = e^{-x} \cdot \sec^2(y) - \ln(z)
$$

Por fim, a derivada parcial de $ W $ em relação a $ z $:

$$
\frac{\partial W}{\partial z} = \frac{\partial}{\partial z} \left( e^{-x} \cdot \tan(y) \right) - \frac{\partial}{\partial z} \left( y \cdot \ln(z) \right)
$$

Não há $ z $ na primeira parte, então essa derivada é zero. A derivada de $ \ln(z) $ em relação a $ z $ é $ \frac{1}{z} $, então a derivada parcial em relação a $ z $ é:

$$
\frac{\partial W}{\partial z} = - \frac{y}{z}
$$

Resumindo, as derivadas parciais são:

$$
\frac{\partial W}{\partial x} = -e^{-x} \cdot \tan(y)
$$

$$
\frac{\partial W}{\partial y} = e^{-x} \cdot \sec^2(y) - \ln(z)
$$

$$
\frac{\partial W}{\partial z} = - \frac{y}{z}
$$


```python
#gerado pelo chatgpt 4.5 direto com o plotly
import numpy as np
import plotly.graph_objs as go
from sympy import symbols, lambdify, tan, ln, exp

# Definir as variáveis simbólicas
x, y, z = symbols('x y z')

# Definir a função W
W = exp(-x) * tan(y) - y * ln(z)

# Lambdify para converter a expressão sympy em uma função que pode ser avaliada numericamente
W_lambdified = lambdify((x, y, z), W, modules=['numpy'])

# Gerar os valores de x e y para o gráfico
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 100)  # Evitar singularidades em tan(y)
X, Y = np.meshgrid(x_vals, y_vals)

# Escolher um valor fixo para z para as curvas de nível
z_val = 1

# Calcular os valores de W para a grade de valores x e y e um valor fixo z
W_vals = W_lambdified(X, Y, z_val)

# Criar a figura 3D
fig = go.Figure(data=[go.Surface(z=W_vals, x=X, y=Y)])

# Adicionar as curvas de nível referentes a z
contours = {
    'z': {
        'show': True,
        'start': W_vals.min(),
        'end': W_vals.max(),
        'size': (W_vals.max() - W_vals.min()) / 20,
        'color':'cyan'
    }
}

fig.update_traces(contours=contours)

# Adicionar títulos e rótulos
fig.update_layout(title='Gráfico 3D de W com z como curva de nível',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='W'),
                  autosize=False,
                  width=800,
                  height=800)

# Mostrar o gráfico
fig.show()
```
