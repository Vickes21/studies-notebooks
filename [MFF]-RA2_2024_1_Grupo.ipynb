###RESPOSTA 1

Temos a função da área $A= x.y$ e a função do perímetro $P=2x + 2y$

Partindo do princípio de que $P=100$, obtemos o valor de $y=\frac{100}{x}$ da função de P.


Tiramos a derivada de $P$, subsituindo o valor de $y$.

$$\frac{d}{dx}(2x + 2\frac{100}{x})$$
$$\frac{d}{dx}(2x + \frac{200}{x})$$
$$\frac{d}{dx}(2 - \frac{200}{x²})$$

Igualamos o perímetro à zero, para encontramos o menor valor possível para $x$:
$$2 - \frac{200}{x²}=0 \therefore x = 10$$


Tiramos a segunda derivada de $P$:

$$\frac{d²}{dx²}P = \frac{d}{dx}(2 - \frac{200}{x²})=\frac{400}{x³}$$

$x=10$ é um ponto crítico (onde $\frac{d}{dx}P = 0$), agora vamos identificar se é um ponto de máximo ou de mínimo, substituindo-o na segunda derivada:

$$\frac{d²}{dx²}f(10) = \frac{400}{10³} = 0.4$$

Encontramos um ponto de mínimo local em $x=10$, pois $0.4$ é maior que zero


Para encontramos o retângulo com menor perímetro precisamos do valor de $y$ que é $\frac{100}{x}$, quando $x=10$;

$$y = \frac{100}{10} = 10$$

Portanto a forma geométrica que procuramos é um quadrado medindo $10$ em cada um de seus lados ($x$ e $y$).




###RESPOSTA 2

1. Relação entre preço e quantidade vendida:
   - A relação base é: para um preço de $1,50$, você vende $5000$ itens.
   - Para cada redução de $0,10$ no preço, a quantidade vendida aumenta em $1000$ unidades.
   - Assim, podemos expressar a quantidade $q$ como uma função do preço $p$:
$$q = 5000 + 10000 \cdot(1,50 - p)$$

2. Função de custo:
   - Os custos fixos são $2000$.
   - O custo marginal por item é $0,50$.
   - O custo total é então:

$$\text{custo_total} = 2000 + 0.50 \cdot q$$

3. Função de receita:
   - A receita é o produto do preço pelo número de itens vendidos:
$$
     \text{receita} = p \cdot q
$$

4. Função de lucro:
   - O lucro é a diferença entre a receita e o custo total:
$$
     \text{lucro} = \text{receita} - \text{custo_total}
$$

5. Maximização do lucro:
   - Para encontrar o preço que maximiza o lucro, derivamos a função de lucro em relação ao preço e igualamos a zero:
$$
     \frac{d(\text{lucro})}{dp} = 0
$$
   - Resolvendo esta equação nos dá o preço ótimo.

```python
from sympy import symbols, diff, solve

# Define as variáveis
p, q = symbols('p q')

# Definir a relação de preço e quantidade vendida
# Quando p = 1.50, q = 5000
# A cada $0.10 de redução, q aumenta em 1000
# Portanto, 1.50 - p = 0.10x --> x = 10 * (1.50 - p)
# q = 5000 + 1000x = 5000 + 10000 * (1.50 - p)
q = 5000 + 10000 * (1.50 - p)

# Custo total = custos fixos + custo marginal por item
# Custo fixo = 2000, custo marginal = 0.50 por item
custo_total = 2000 + 0.50 * q

# Receita = preço * quantidade
receita = p * q

# Lucro = Receita - Custo total
lucro = receita - custo_total

# Derivada do lucro em relação ao preço
dlucro_dp = diff(lucro, p)

# Encontrar o preço que maximiza o lucro (zerar a derivada)
preco_otimo = solve(dlucro_dp, p)
preco_otimo
```

###Quantidade Vendida vs. Preço:

À medida que o preço diminui de $1,50$ para $0,50$, a quantidade vendida aumenta. Isso é mostrado pela curva ascendente no primeiro gráfico. A linha vermelha pontilhada indica o preço ótimo de $1,25$, onde a maximização do lucro é teoricamente alcançada.

###Lucro vs. Preço:
O segundo gráfico mostra o lucro em função do preço. O lucro aumenta inicialmente com a redução do preço devido ao aumento significativo na quantidade vendida, superando o decréscimo no preço por unidade. No entanto, após atingir um ponto máximo (indicado pela linha vermelha pontilhada em $1,25$), o lucro começa a cair porque o preço baixo não compensa o aumento nos custos variáveis e a maior quantidade vendida.  

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de preços
precos = np.linspace(0.50, 1.50, 100)

# Calculando as quantidades correspondentes
quantidades = 5000 + 10000 * (1.50 - precos)

# Calculando os custos totais
custos_totais = 2000 + 0.50 * quantidades

# Calculando as receitas
receitas = precos * quantidades

# Calculando os lucros
lucros = receitas - custos_totais

# Criando os gráficos
plt.figure(figsize=(14, 6))

# Gráfico de quantidade vs. preço
plt.subplot(1, 2, 1)
plt.plot(precos, quantidades, label='Quantidade Vendida')
plt.title('Quantidade Vendida vs. Preço')
plt.xlabel('Preço')
plt.ylabel('Quantidade Vendida')
plt.grid(True)
plt.axvline(x=1.25, color='r', linestyle='--', label='Preço Ótimo (1,25)')
plt.legend()

# Gráfico de lucro vs. preço
plt.subplot(1, 2, 2)
plt.plot(precos, lucros, label='Lucro')
plt.title('Lucro vs. Preço')
plt.xlabel('Preço')
plt.ylabel('Lucro')
plt.grid(True)
plt.axvline(x=1.25, color='r', linestyle='--', label='Preço Ótimo (1,25)')
plt.legend()

plt.tight_layout()
plt.show()
```
